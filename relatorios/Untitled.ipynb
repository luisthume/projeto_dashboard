```python
import pandas as pd
```

```python
def parseCNPJ(x):
    try:
        return x.replace('.', '').replace('/', '').replace('-','')
    except AttributeError:
        pass
```

```python
clientes_fornecedores = pd.read_csv('clientes_fornecedores.csv', error_bad_lines=False, sep=';')
clientes_fornecedores.rename(columns={'fornecedor':'empresa','document':'documento'}, inplace=True)
clientes_fornecedores.documento = clientes_fornecedores.documento.apply(lambda x: parseCNPJ(x))
clientes_fornecedores.set_index('documento', inplace=True)
```
